```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
spotify = pd.read_csv("C:\\mydata\\spotify.csv")
spotify.head()
```

```python
spotify = pd.read_csv("C:\\mydata\\spotify.csv", index_col = "Date")
spotify.head()
```

```python
plt.figure(figsize = (20, 8))
sns.lineplot(data = spotify)
plt.show()
```
