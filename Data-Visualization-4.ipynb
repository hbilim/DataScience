```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
canada = pd.read_csv("C:\\mydata\\canada1.csv", delimiter = ';', encoding='windows-1252')
canada.head(15)
```

```python
canada.columns
```

```python
canada = canada.drop(columns = ['AREA', 'DEV', 'DevName', 'REG', 'Type', 'Coverage', 'AreaName', 'RegName'])
canada.head()
```

```python
canada.rename(columns ={'OdName':'Country'}, inplace = True)
canada.set_index(canada.Country, inplace = True)
canada.head()
```

```python
canada.index.name = None
canada.head()
```

```python
del canada['Country']
canada.head()
```

```python
canada = canada.transpose()
```

```python
canada.head()
```

```python
plt.figure(figsize = (18, 8))
plt.title('Immigrants from Uruguay')
sns.set(rc = {"axes.facecolor" : '#B0C4DE', "axes.grid" : False, 'xtick.labelsize' : 10, 'ytick.labelsize' : 10})
plt.xticks(rotation = 45)
sns.lineplot (x = canada.index.values, y = canada['Uruguay'])
plt.show()
```

```python
plt.figure(figsize = (18, 8))
plt.title('Immigrants from Finland, China and Zambia')
sns.set(rc = {"axes.facecolor" : '#283747', 'axes.grid' : False, 'xtick.labelsize' : 15, 'ytick.labelsize' : 15})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['Finland'], color = '#FF9900', label = 'Finland')
sns.lineplot(x = canada.index.values, y = canada['China'], color = '#4586FF', label = 'China')
sns.lineplot(x = canada.index.values, y = canada['Zambia'], color = '#A2EF44', label = 'Zambia')
plt.legend(facecolor = 'grey', fontsize = 'large', edgecolor = 'black', shadow = True)
plt.show()
```
