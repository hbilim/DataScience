```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("C:/mydata/2013-2023-yl-ibb-calisanlar-orenim.csv",delimiter = ';', encoding='windows-1252')
data.head(10)
```

```python
plt.figure(figsize = (7, 7))
sns.scatterplot(x = data['Yil'], y = data['Lise Kadin'], color = 'red')
sns.scatterplot(x = data['Yil'], y = data['Lise Erkek'], color = 'blue')
plt.show()
```

```python
plt.figure(figsize = (7, 7))
plt.title('Graduated Women vs Men')
sns.scatterplot(x = data['Yil'], y = data['Lisans Kadin'], color = '#91bd3a', marker = '>', label = 'Women')
sns.scatterplot(x = data['Yil'], y = data['Lisans Erkek'], color = '#A52A2A', marker = 'v', label = 'Men')
plt.xlabel('Years')
plt.ylabel('Number of People')
plt.show()
```

```python
plt.figure(figsize = (7, 7))
plt.title('Literacy Women vs Men')
sns.scatterplot(x = data['Yil'], y = data['Okuryazar Kadin'], color = '#91bd3a', label = 'Women')
sns.scatterplot(x = data['Yil'], y = data['Okuryazar Erkek'], color = '#A52A2A', label = 'Men')
plt.xlabel('Years')
plt.ylabel('Number of People')
plt.show()
```
