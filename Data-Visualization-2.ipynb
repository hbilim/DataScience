```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import warnings
warnings.filterwarnings('ignore')
```

```python
#SEABORN
```

```python
col1 = np.linspace(0, 50, 1000)
col2 = np.sin(col1)
df = pd.DataFrame({"C1": col1, "C2":col2})
df.head(10)
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
sns.lineplot(x = df.C1, y= df.C2, data = df)
plt.show()
```

```python
col1 = np.linspace(0, 20, 1000)
col2 = np.sin(col1)
df = pd.DataFrame({"C1": col1, "C2":col2})
df.head(10)
```

```python
sns.lineplot(x = df.C1, y = df.C2, data = df)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.set(rc = {"axes.facecolor" : "#FAFAD2", "axes.grid" : True, "xtick.labelsize" : 14, "ytick.labelsize" : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 2)
sns.lineplot(x = df.C1, y= df.C2 + 1, data = df, linewidth = 2)
plt.show()
```
